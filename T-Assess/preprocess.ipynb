```python
import os
 
path = '../datasets/rome/rome'
files = [f for f in os.listdir(path) if f.endswith('.txt')]
 
#保存到merged_file.txt文件中
with open('../datasets/rome/merged_file.txt', 'w') as outfile:
    for file in files:
        with open(os.path.join(path, file)) as infile:
            outfile.write(infile.read() + '\n')
```

```python
import pandas as pd
gps_data = pd.read_csv('../datasets/rome/merged_file.txt',names=['tid','time','longitude','latitude'])
```

```python
gps_data
```

```python
gps_data=gps_data.sort_values(by=['tid','time']).reset_index(drop=True)
```

```python
gps_data
```

```python
print("数据集共有", len(gps_data), "条gps数据")
```

```python
# 数据去重
gps_data_drop_dup = gps_data.drop_duplicates().reset_index(drop=True)
```

```python
print("去重后数据集共有", len(gps_data_drop_dup), "条gps数据")
```

```python
gps_data_drop_dup['pid'] = gps_data_drop_dup.index 
gps_data_drop_dup['pid'] = gps_data_drop_dup['pid'] + 1
order = ['time', 'longitude', 'latitude', 'tid', 'pid']
gps_data_drop_dup = gps_data_drop_dup[order]
```

```python
gps_data_drop_dup
```

```python
#保存数据清理之后的gps数据
gps_data_drop_dup.to_csv('../trajectory-quality-assessment/data/rome_preprocess/rome.txt', header=0, index=False)
```
